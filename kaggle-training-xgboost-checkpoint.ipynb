## ğŸ“‚ Step 4: Locate and Verify Checkpoint Embeddings

```python
# TÃ¬m checkpoint trong Kaggle input datasets
import os
from pathlib import Path

checkpoint_found = False
checkpoint_path = None

if Path('/kaggle/input').exists():
    print("ğŸ” TÃ¬m checkpoint trong /kaggle/input/...")
    print("="*70)
    
    for input_dir in Path('/kaggle/input').iterdir():
        if input_dir.is_dir():
            potential_checkpoint = input_dir / 'checkpoints'
            checkpoint_state = potential_checkpoint / 'checkpoint_state.json'
            
            if checkpoint_state.exists():
                print(f"âœ… TÃ¬m tháº¥y checkpoint: {potential_checkpoint}")
                checkpoint_path = str(potential_checkpoint)
                checkpoint_found = True
                
                # List files
                print("\nğŸ“ Ná»™i dung checkpoint:")
                for file in sorted(potential_checkpoint.iterdir()):
                    if file.is_file():
                        size_mb = file.stat().st_size / 1024 / 1024
                        print(f"  - {file.name:50s} ({size_mb:8.2f} MB)")
                break
    
    if not checkpoint_found:
        print("âŒ KhÃ´ng tÃ¬m tháº¥y checkpoint!")
        print("   HÃ£y Ä‘áº£m báº£o báº¡n Ä‘Ã£ thÃªm checkpoint dataset vÃ o Kaggle notebook.")
else:
    print("âš ï¸  KhÃ´ng cháº¡y trÃªn Kaggle hoáº·c khÃ´ng cÃ³ /kaggle/input")

print("="*70)
```

```python
# Kiá»ƒm tra checkpoint state
if checkpoint_found:
    import json
    
    with open(Path(checkpoint_path) / 'checkpoint_state.json') as f:
        state = json.load(f)
    
    print("ğŸ“Š CHECKPOINT STATUS")
    print("="*70)
    print(f"Last completed stage: {state['last_completed_stage']}")
    print(f"\nCompleted stages ({len(state['completed_stages'])}):{"")}")
    for stage in state['completed_stages']:
        metadata = state['metadata'].get(stage, {})
        print(f"  âœ“ {stage:30s} {str(metadata)}")
    print("="*70)
else:
    print("âŒ KhÃ´ng thá»ƒ kiá»ƒm tra checkpoint state (checkpoint khÃ´ng tá»“n táº¡i)")
```

## ğŸ“ˆ Step 6: Check Results

```python
# Hiá»ƒn thá»‹ grid search results
import json
from pathlib import Path

gridsearch_results = Path('model_phase/results/gridsearch_xgboost/gridsearch_results.json')

if gridsearch_results.exists():
    with open(gridsearch_results) as f:
        results = json.load(f)
    
    print("ğŸ” GRID SEARCH RESULTS")
    print("="*80)
    print(f"Tested {len(results['results'])} configurations\n")
    
    # Sort by F1 score
    sorted_results = sorted(results['results'], key=lambda x: x['val_f1'], reverse=True)
    
    print("Top 5 Configurations:")
    print("-"*80)
    for i, config in enumerate(sorted_results[:5], 1):
        print(f"{i}. n_est={config['n_estimators']:3d}, depth={config['max_depth']:2d}, lr={config['learning_rate']:.2f} â†’ "
              f"F1={config['val_f1']:.4f}, Acc={config['val_accuracy']:.4f}")
    
    print("\n" + "="*80)
    print("ğŸ† BEST CONFIGURATION:")
    best = results['best_config']
    print(f"  n_estimators: {best['n_estimators']}")
    print(f"  max_depth: {best['max_depth']}")
    print(f"  learning_rate: {best['learning_rate']}")
    print(f"  Val F1: {best['val_f1']:.4f}")
    print(f"  Val Accuracy: {best['val_accuracy']:.4f}")
    print("="*80)
else:
    print("âš ï¸  Grid search results chÆ°a cÃ³.")
```

```python
# Hiá»ƒn thá»‹ final model results
import os

# TÃ¬m thÆ° má»¥c káº¿t quáº£ má»›i nháº¥t
results_dir = Path('model_phase/results')
xgboost_dirs = [d for d in results_dir.iterdir() if d.is_dir() and d.name.startswith('xgboost_')]

if xgboost_dirs:
    latest_dir = max(xgboost_dirs, key=lambda x: x.stat().st_mtime)
    results_file = latest_dir / 'results.json'
    
    if results_file.exists():
        with open(results_file) as f:
            final_results = json.load(f)
        
        print("ğŸ“Š FINAL MODEL RESULTS")
        print("="*80)
        print(f"Model: {final_results['model_config']['model_name']}")
        print(f"Embedding: {final_results['model_config']['embedding_model']}")
        print(f"\nHyperparameters:")
        print(f"  n_estimators: {final_results['model_config']['n_estimators']}")
        print(f"  max_depth: {final_results['model_config']['max_depth']}")
        print(f"  learning_rate: {final_results['model_config']['learning_rate']}")
        
        print(f"\nDataset:")
        print(f"  Train: {final_results['dataset_info']['train_size']:,} samples")
        print(f"  Val: {final_results['dataset_info']['val_size']:,} samples")
        print(f"  Test: {final_results['dataset_info']['test_size']:,} samples")
        
        print(f"\nValidation Performance:")
        print(f"  Accuracy: {final_results['validation_accuracy']:.4f}")
        print(f"  Precision: {final_results['validation_precision']:.4f}")
        print(f"  Recall: {final_results['validation_recall']:.4f}")
        print(f"  F1-score: {final_results['validation_f1']:.4f}")
        
        if final_results['dataset_info']['test_size'] > 0:
            print(f"\nTest Performance:")
            print(f"  Accuracy: {final_results['test_accuracy']:.4f}")
            print(f"  Precision: {final_results['test_precision']:.4f}")
            print(f"  Recall: {final_results['test_recall']:.4f}")
            print(f"  F1-score: {final_results['test_f1']:.4f}")
        
        print(f"\nTraining Time: {final_results['training_time']:.2f}s")
        print("="*80)
        print(f"\nâœ… Model saved to: {latest_dir}")
    else:
        print("âš ï¸  Final results file not found.")
else:
    print("âš ï¸  No XGBoost results directories found.")
```
